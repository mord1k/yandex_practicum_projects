### Проверка правильности расчёта эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака `rougher.output.recovery`.

Создадим переменную `recovery_output`, в которой будут храниться данные признака эффективности обогащения из обучающей выборки. 

```python
recovery_output = data_train['rougher.output.recovery']
```

Для проверки правильности расчёта эффективности обогащения необходимо использовать формулу:

**Recovery = (C * (F-T)) / (F * (C-T)) * 100**,

где 
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Для начала найдем в датасете все необходимые признаки. Для переменной `C` необходимо найти признак `rougher.output.concentrate_au`.

```python
C = data_train['rougher.output.concentrate_au']
```

Переменная `F` должна быть для признака `rougher.input.feed_au`.

```python
F = data_train['rougher.input.feed_au']
```

Переменная `T` будет соответсвовать признаку `rougher.output.tail_au'`.

```python
T = data_train['rougher.output.tail_au']
```

Создадим отдельный датафрейм для самостоятельного расчёта эффективности обогащения. Назовем его переменной `recovery`. Применим формулу, которая описана выше.

```python
recovery = pd.DataFrame()
```

```python
recovery = (C * (F-T)) / (F * (C-T)) * 100
```

Сравним данные в `recovery` и `recovey_output`.

```python
recovery_output.head()
```

```python
recovery.head()
```

На первый взгялд данные похожи. Посмотрим на метрику качества — `MAE` (англ. mean absolute error). Среднее абсолютное отклонение расчитаем с помощью функции `mean_absolute_error`.

```python
mean_absolute_error(recovery, recovery_output)
```

После расчёта метрики `MAE` можно сделать вывод, что проверка на эффективность обогащения рассчитана правильно. Показатели `MAE` слишком маленькие, и говорят нам о том, что разницы в данных нет.

## Модель

### Функция sMAPE

Напишем функцию для вычисления итоговой sMAPE.

Для решения задачи введём новую метрику качества — `sMAPE` (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

```python
def smape(target, predicted):
    smape = 100/len(target) * np.sum(2 * np.abs(target - predicted) / (np.abs(target) + np.abs(predicted)))
    return smape
```

Для вычисления итоговой sMAPE будем использовать функцию.

```python
def total_smape(rougher, final):
    return 0.25 * rougher + 0.75 * final
```

Для обучения моеделей на кросс-валидации необходимо использовать функцию `make_scorer`.

```python
smape_score = make_scorer(smape, greater_is_better=False)
```
